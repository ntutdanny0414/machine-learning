```python
#載入資料
import pandas as pd  
import numpy as np  
data = pd.read_csv('iris.csv')
data
```

```python
sepal_length = data["sepal_length"].values
```

```python
max(sepal_length),min(sepal_length)
```

資料分布

```python
import matplotlib.pyplot as plt
x1 = data[data["species"] == "setosa"]["sepal_length"].values
x2 = data[data["species"] == "versicolor"]["sepal_length"].values
x3 = data[data["species"] == "virginica"]["sepal_length"].values
kwargs = dict(histtype='stepfilled',alpha=0.3,bins=100,range = [min(sepal_length),max(sepal_length)])


plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.show()
```

histogram estimator plot

```python
hist,bins = np.histogram(sepal_length,bins = 100)
```

```python
hist,bins
```

```python
import matplotlib.pyplot as plt
plt.hist(sepal_length,bins=bins,range = [min(sepal_length),max(sepal_length)])
plt.show()
```

```python
def px_compute(x):
    mask1 = data["sepal_length"] <= x + 0.018
    mask2 = data["sepal_length"] > x - 0.018
    return len(data[mask1&mask2])/(150*0.036)
```

```python
N = 150 #Point
h = 0.036
p_x_index = []
p_x = []
for i in range(len(bins)-1):#建index
    for j in range(10):#at least 10 point
        p_x_index.append(bins[i]+(j*(h/10)))
p_x_index.append(7.9)
for i in range(len(p_x_index)):
    p_x.append(px_compute(p_x_index[i]))
p_x
```

結果

```python
from matplotlib import pyplot as plt 
plt.title("Histogram Naive estimator") 
plt.xlabel("length") 
plt.ylabel("p_x") 
plt.plot(p_x_index,p_x) 
#plt.savefig('Q5_MLE_outcome.png')
plt.show()
```
