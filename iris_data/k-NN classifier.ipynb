#### k-NN classifier

```python
#載入資料
import pandas as pd  
import numpy as np  
data = pd.read_csv('iris.csv')
data
```

```python
#Accuracy
def Accuracy(y_pred,y_test):
    count = 0
    for i in range(len(y_pred)):
        if y_pred[i] == y_test[i]:
            count = count + 1
    return count / len(y_pred)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

#split data and train
X = data.iloc[:,:4].values
y = data['species'].replace({'setosa':0,'versicolor':1,'virginica':2}).copy().values
K = []
Accuracy_with_K = []
```

```python
#train
for k in range(1,16,2):
    Accuracy_list = []
    for i in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)#random split
        KNNclassifier = KNeighborsClassifier(n_neighbors=k,p = 2)#p=2 Euclidean
        KNNclassifier.fit(X_train, y_train)
        y_pred = KNNclassifier.predict(X_test)
        Accuracy_list.append(Accuracy(y_pred,y_test))
    K.append(k)
    Accuracy_with_K.append(sum(Accuracy_list) / 10)
print ('Average accuracy with K(1,3,..,15):', Accuracy_with_K)
```

```python
#plot
from matplotlib import pyplot as plt 
plt.title("KNN Classifier Outcome") 
plt.xlabel("K") 
plt.ylabel("accuracy") 
plt.plot(K,Accuracy_with_K) 
#plt.savefig('KNN_Classifier_Outcome.png')
plt.show()
```
