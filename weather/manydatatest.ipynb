```python
import pandas as pd  
import numpy as np  
from sklearn.metrics import explained_variance_score, mean_absolute_error,median_absolute_error
from sklearn.model_selection import train_test_split  
from sklearn import preprocessing,svm
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('inputdata.csv')
df
```

```python
df['風向'].fillna('無資料', inplace=True)
df['風向'].unique()
```

```python
dictwind = {'北':0,'北北東':22.5,'東北':45,'東北東':68,'東':90,'東南東':112.5,'東南':135,'南南東':158,'南':180, 
            '南南西':202.5,'西南':225,'西南西':248,'西':270,'西北西':293,'西北':315,'北北西':338,
            '靜風':360,'風向不定':360,'無資料':360}
```

```python
df['風向'] = df['風向'].map(dictwind)
```

```python
df
```

```python
df.set_index('監測時間')
numattr = ['測站氣壓(hPa)','監測日期','監測時間','溫度(℃)','相對溼度(%)','風向','風速(m/s)']
df = df[numattr]
feature = '溫度(℃)'
for i in range(0,48360):
#    df.iloc[i:i,'....']
    df.iloc[i,1] = df.iloc[i,1].split('/')[1]
    df.iloc[i,2] = df.iloc[i,2].split(':')[0]
    df.iloc[i,0] = df.iloc[i,0].replace(',','')
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
df.to_csv('manycleandata.csv',encoding='utf_8_sig',index=False)
```
