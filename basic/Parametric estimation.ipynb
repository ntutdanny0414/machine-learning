```python
def Likelihood_function(Theta):
    "Example function"
    return Theta * Theta * (1 - Theta) * (1 - Theta) * Theta
```

```python
Theta_point = []
probability = []

"建立Theta_point"
for i in range(0,201):
    Theta_point.append(i)
for i in range(len(Theta_point)):
    Theta_point[i] = Theta_point[i] / 200

"Likelihood_function"
for Theta in Theta_point:
    probability.append(Likelihood_function(Theta))
```

```python
from matplotlib import pyplot as plt 
"plot"
plt.title("MLE example") 
plt.xlabel("Theta") 
plt.ylabel("probability") 
plt.plot(Theta_point,probability) 
#plt.savefig('Q5_MLE_outcome.png')
plt.show()
```

```python
"MAX Theta"
MAX_probability  = max(probability) #最大機率
MAX_Theta = Theta_point[probability.index(MAX_probability)] #最大Theta
print(MAX_Theta) #為0.6沒錯
```
