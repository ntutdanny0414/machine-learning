# Step 6 Use matplotlib show 2D images about data.
[Matplotlib](https://matplotlib.org/)
### Plot the distribution of two class (1. GDP < 58000, 2. GDP >= 58000) of COVID-19 State Data using different colors and different  marker where x-axis is the Pollution and y-axis the Mortality-rate.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df['Mortality-rate'] = df['Deaths'] / df['Infected']
df_small = df[df['GDP'] < 58000]
df_large = df[df['GDP'] >= 58000]
```

```python
plt.xlabel('Pollution')
plt.ylabel('Mortality-rate')
plt.scatter(df_small['Pollution'], df_small['Mortality-rate'], c = 'red',marker ='o',alpha = 0.8,label='GDP < 58000')
plt.scatter(df_large['Pollution'], df_large['Mortality-rate'], c = 'blue',marker ='x',alpha = 0.5,label='GDP >= 58000')
plt.legend(loc='upper right')
plt.show()
```

# Step 7
### Show the proportion of three class of COVID-19 State Data using pie chart.

**<font color = blue size=3> About the class: </font>** <p>
 1、 Mortality-rate < 0.02 <p>
 2、 Mortality-rate between 0.02 and 0.03 <p>
 3、Mortality-rate > 0.03 <p>

```python
mask1 = df['Mortality-rate'] < 0.02
mask2_1 = df['Mortality-rate'] >= 0.02
mask2_2 = df['Mortality-rate'] <= 0.03
mask3 = df['Mortality-rate'] > 0.03
plt.pie(
    [len(df[mask1]), len(df[mask2_1 & mask2_2]), len(df[mask3])],
    labels = ['Mortality-rate < 0.02', 'Mortality-rate between 0.02 and 0.03', 'Mortality-rate > 0.03'],
    autopct= "%3.1f%%")
plt.show()
```
