```python
import pandas as pd 
import numpy as np
data_list = []
for i in range(8):
    data_list.append(pd.read_csv('HW2_traffic_accident/'+str(101+i)+'Traffic_accident.csv'))
```

```python
def plot_info(df):
    info_1_1 = list(df.groupby('Month').sum().sort_values(by=['death_toll'],ascending = False).index)
    info_1_2 = list(df.groupby('Month').sum().sort_values(by=['death_toll'],ascending = False)['death_toll'])
    info_2_1 = list(df.groupby('Month').sum().sort_values(by=['Injuries_Person'],ascending = False).index)
    info_2_2 = list(df.groupby('Month').sum().sort_values(by=['Injuries_Person'],ascending = False)['Injuries_Person'])
    info_3_1 = list(df.groupby('Vehicle_type').sum().sort_values(by=['Injuries_Person'],ascending = False).head(10).index)
    info_3_2 = list(df.groupby('Vehicle_type').sum().sort_values(by=['Injuries_Person'],ascending = False).head(10)['Injuries_Person'])
    info_4_1 = list(df.groupby('Weather').sum().sort_values(by=['Injuries_Person'],ascending = False).index)
    info_4_2 = list(df.groupby('Weather').sum().sort_values(by=['Injuries_Person'],ascending = False)['Injuries_Person'])
    info_5_1 = list(df.groupby('Road_pattern').sum().sort_values(by=['Injuries_Person'],ascending = False).head(10).index)
    info_5_2 = list(df.groupby('Road_pattern').sum().sort_values(by=['Injuries_Person'],ascending = False).head(10)['Injuries_Person'])
    info_6_1 = list(df.groupby('Accident_location').sum().sort_values(by=['Injuries_Person'],ascending = False).head(10).index)
    info_6_2 = list(df.groupby('Accident_location').sum().sort_values(by=['Injuries_Person'],ascending = False).head(10)['Injuries_Person'])
    return [(info_1_1,info_1_2),(info_2_1,info_2_2),(info_3_1,info_3_2),(info_4_1,info_4_2),(info_5_1,info_5_2),(info_6_1,info_6_2)]
```

```python
year = []
for i in range(len(data_list)):
    year.append(plot_info(data_list[i]))   
```

```python
x_label_list = ['Month','Month','Vehicle_type','Weather','Road_pattern','Accident_location']
y_label_list = ['death_toll','Injuries_Person','Injuries_Person','Injuries_Person','Injuries_Person','Injuries_Person']
title_list = ['Month','Month','TOP 10 Vehicle_type','Weather','TOP 10 Road_pattern','TOP 10 Accident_location']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```
